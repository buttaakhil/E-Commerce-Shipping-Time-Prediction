```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("Shipment.csv")
```

```python
dataset.head()
```

```python
dataset.drop(columns=["ID","Gender"],inplace=True)
```

```python

dataset["Reached.on.Time_Y.N"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
le=LabelEncoder()
dataset["Warehouse_block"]=pd.DataFrame(le.fit_transform(dataset["Warehouse_block"]))
dataset["Mode_of_Shipment"]=pd.DataFrame(le.fit_transform(dataset["Mode_of_Shipment"]))
```

```python
ord_data=[["low","medium","high"]]
oe=OrdinalEncoder(categories=ord_data)
dataset["Product_importance"]=pd.DataFrame(oe.fit_transform(dataset[["Product_importance"]]))
dataset["Product_importance"]=dataset["Product_importance"].astype("int64")
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Reached.on.Time_Y.N"]
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru=RandomUnderSampler()
x,y=ru.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x)
x=pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel="rbf",C=1,gamma=0.1)
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train)*100,sv.score(x_test,y_test)*100
```
